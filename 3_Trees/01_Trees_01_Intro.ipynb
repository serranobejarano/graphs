# Árboles

Un árbol es un grafo conexo que no tiene ciclos. A los vértices de grado 1 se le llaman hojas. 
Los conjuntos de árboles que no están conectados entre ellos se les llama bosque.


Sea G un grafo no dirigido simple, es decir, sin bucles, se cumple que:



- Un grafo no dirigido es un árbol si hay un único camino entre cada pareja de vértices.
- Un árbol en el que el número de aristas es mayor o igual a 2 tiene siempre al menos dos hojas (vértices de grado 1)
- Un árbol con n vértices tiene n-1 aristas.


# Árboles binarios

- Cuando el grado de salida de cada vértice es 0 ó 2.
- De la misma forma en un árbol N-ario, cada vértice tiene 0 o N hijos.
- En un árbol N-ario completo o balanceado cada vértice tiene N hijos en cada nivel salvo el último.


```python
bi_tree = nx.balanced_tree(2,5, create_using=nx.DiGraph)
```

```python
bi_tree.nodes
```

```python
#pos = nx.drawing.nx_agraph.graphviz_layout(bi_tree, prog="dot")
pos = nx.layout.kamada_kawai_layout(bi_tree)
nx.draw(bi_tree, pos=pos)
```

```python
#pos = nx.drawing.nx_agraph.graphviz_layout(bi_tree, prog="neato")
pos = nx.layout.kamada_kawai_layout(bi_tree)
nx.draw(bi_tree, pos=pos)
```

# Arboles balanceados


Un árbol balanceado es un árbol binario en el cual las alturas de dos subárboles de todo vértice se diferencia en 1 como máximo. El blance de un vértice se define como la altura de su subárbol izquierdo menos la altura de su subárbol derecho.

```python
b_tree = nx.balanced_tree(3,3, create_using=nx.DiGraph)
```

```python
nx.draw_kamada_kawai(b_tree)
```

```python
nx.draw_kamada_kawai(b_tree.to_undirected())
```

# Arboles r-ary (k-ary o n-ary)

Los árboles-R o R-árboles son estructuras de datos de tipo árbol similares a los árboles-B, con la diferencia de que se utilizan para métodos de acceso espacial, es decir, para indexar información multidimensional; por ejemplo, las coordenadas (x, y) de un lugar geográfico. Un problema con aplicación práctica en el mundo real podría ser: "Encontrar todos los museos en un radio de dos kilómetros alrededor de la posición actual".

```python
rary_tree = nx.generators.classic.full_rary_tree(r=5,n=20,create_using=nx.DiGraph)
```

```python
nx.draw_kamada_kawai(rary_tree)
```

```python
nx.draw_kamada_kawai(rary_tree.to_undirected())
```

## Spanning trees

Un árbol de expansión de un grafo conexo, no dirigido G es un árbol compuesto por todos los vértices y algunas (quizá todas) de las aristas de G. Informalmente, un árbol de expansión de G es una selección de aristas de G que forman un árbol que cubre todos los vértices. 

```python
min_s_tree = nx.tree.minimum_spanning_tree(G)
```

```python
nx.draw(min_s_tree,seed=np.random.seed(42), with_labels=True)
```

# Branchings y Arborescencias optimos a partir de un Grafo

```python
edmonds = nx.tree.Edmonds(G)
edmonds_optimum = edmonds.find_optimum()
nx.draw(edmonds_optimum,seed=np.random.seed(42), with_labels=True)
```
